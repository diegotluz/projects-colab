```python
from pytube import YouTube
lnk='https://www.youtube.com/watch?v=dT_aIJZ7Mxw'
yt = YouTube(lnk)
listtags= yt.streams.filter(file_extension='mp4').get_audio_only()
print(listtags)
# stream = yt.streams.get_by_itag(18)
listtags.download('',f'{yt.title}.mp4')
```

```python
import whisper
model = whisper.load_model("base")
result = model.transcribe(f"{yt.title}.mp4")
print(result['text'])
```
